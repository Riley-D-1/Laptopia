```python
import pandas as pd
df = pd.read_csv("data/scraped_data.csv")
print(df)
```

```python
print(df.columns.tolist())
rename_stuff = {
    "Graphics processor": "Graphics",
    "Display size (inches)": "Screen Size (Inches)",
    "Processor Cores": "Core Count",
    "Processor Type": "Processor",
    "Processor Clock Speed (GHz)": "Clock Speed (GHz)",
    "Operating system": "OS",
    "Product Weight (kg)": "Weight (kg)",
    "Operating system": "OS",
    "Manufacturer's warranty": "Warranty",
    "Total Storage": "Storage",
    "Manufacturer's warranty": "Warranty",
    "Manufacturer's Warranty": "Warranty",
    "Price_AUD": "Price (AUD)",
}
df.rename(columns=rename_stuff,inplace=True)
df["Brand"] = df["Product Name"].str.split().str[0]
df["Product Name"] = df["Product Name"].str.split(" ").str[1:].str.join(" ")
needed_vals = [
    "Brand","Product Name","Screen Size (Inches)","Display type",
    "Resolution (Pixels)","Processor","Clock Speed (GHz)","Core Count",
    "Graphics","RAM (GB)","RAM type","Storage","Wi-Fi","OS","Weight (kg)",
    "Warranty","Price (AUD)","Link"
]
cleaned_df = df[needed_vals]
cleaned_df.to_csv("data/cleaned_scraped_data.csv", index=True)
```

```python
df1 = pd.read_csv("data/handmade_data.csv")
df2 = pd.read_csv("data/cleaned_scraped_data.csv")
final_df = pd.concat([df1, df2], ignore_index=True)
final_df["Product Name"] = final_df["Product Name"].str.replace('"', '', regex=False)
final_df.drop(columns=["Warranty.1"], inplace=True)
final_df.drop(columns=["Unnamed: 0"], inplace=True)
final_df.to_csv("data/final_dataset.csv", index=True)
```
